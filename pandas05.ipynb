```python
df = pd.read_csv("cars_tratado.csv", index_col=0)
df.drop(columns='index', inplace=True)
dict_num_of_doors = {
    2: 'two',
    4: 'four'
}
df['num-of-doors-numeric'] = df['num-of-doors']
df['num-of-doors'] = df['num-of-doors'].map(dict_num_of_doors)
```

```python
# Maneira 1

plt.scatter(df['price'], df['horsepower'], color='red')

plt.title("Preo x Horsepower")
plt.xlabel("Preo")
plt.ylabel("Horsepower")

plt.show()
```

```python
# Maneira 2

#plt.scatter(df['price'], df['horsepower'], color='red')
df.plot(kind='scatter', x='price', y='horsepower')

plt.title("Preo x Horsepower")
plt.xlabel("Preo")
plt.ylabel("Horsepower")

plt.show()
```

```python
df.plot(
    kind='scatter', 
    x='price', 
    y='horsepower',
    figsize=(8, 4),
    s=80,
    alpha=0.5,
    color='green')

plt.title("Preo x Horsepower")
plt.xlabel("Preo")
plt.ylabel("Horsepower")

plt.show()
```

#### Subplots

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))


fig.suptitle("Titulo Geral")
axes[0].set_title("Frequencia Preo")
axes[1].set_title("Preo x Horsepower")


df['price'].plot(kind='hist', ax=axes[1])

df.plot(
    kind='scatter', 
    x='price', 
    y='horsepower',
    s=80,
    alpha=0.5,
    color='green',
    ax=axes[0])

plt.show()
```
