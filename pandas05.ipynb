```python
## Formula do Preo

price = 166.86001569 * engine_size  + (-7963.338906281042)
```

```python
df['new_price'] =  166.86001569 * df['engine-size'] + (-7963.338906281042)

yhat = lr.predict(x)
df['new_price_hat'] = yhat
```

```python
df
```

```python
from sklearn import metrics
```

```python
y
```

```python
yhat
```

```python
mae = metrics.mean_absolute_error(y, yhat)
mae
```

```python
mse = metrics.mean_squared_error(y, yhat)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
r2 = metrics.r2_score(y, yhat)
r2 * 100
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['engine-size']]
y = df['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3) #holdout

lr.fit(x_train, y_train)
yhat = lr.predict(x_test)
```

```python
mae = metrics.mean_absolute_error(y_test, yhat)
mae
```

```python
mse = metrics.mean_squared_error(y_test, yhat)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
r2 = metrics.r2_score(y_test, yhat)
r2 * 100
```
