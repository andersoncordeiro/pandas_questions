```python
import pandas as pd
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("cars_tratado.csv", index_col=0)
```

```python
df.drop(columns='index', inplace=True)
```

```python
dict_num_of_doors = {
    2: 'two',
    4: 'four'
}
```

```python
df['num-of-doors-numeric'] = df['num-of-doors']
df['num-of-doors'] = df['num-of-doors'].map(dict_num_of_doors)
```

```python
df['body-style'].unique()
```

```python
df
```

```python
frequencia = df['fuel-type'].value_counts()
percentual = df['fuel-type'].value_counts(normalize=True) * 100
```

```python
dist_freq_qualitativa = pd.DataFrame({
    "Frequencia" : frequencia,
    "Porcentagem": percentual  
})
```

```python
dist_freq_qualitativa.rename_axis("fuel-type", axis=1)
```

### Crosstab

```python
pd.crosstab(df['fuel-type'], df['num-of-doors'])
```

```python
pd.crosstab(df['fuel-type'], df['num-of-doors'], normalize=True) * 100
```

```python
df.groupby('fuel-type')['num-of-doors'].value_counts(normalize=True)  * 100
```

```python
pd.crosstab(df['fuel-type'], df['body-style'], normalize=True, margins=True) * 100
```

```python
pd.crosstab(df['fuel-type'], df['body-style'])
```

```python
pd.crosstab(df['fuel-type'], 
            df['body-style'], 
            aggfunc = ['mean', 'min', 'max'],
            values = df['price']).T
```

```python
df.groupby(by=['fuel-type', 'body-style'])['price'].mean()
```

```python
df_medias_preco = df.groupby(by=['fuel-type', 'body-style'])['price'].mean().reset_index()
```
